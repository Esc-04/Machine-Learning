✔ 결론: 판다스는 “실제 데이터만 열로 친다”시리즈는 1개의 열col 이다.

열 = 데이터(values)만!

index = row, 그냥 붙어 있는 이름표(행 라벨)

```python
sr=pd.Series([30900,22500,2000,3000],index=["피자","치킨","콜라","맥주"])
print(sr.values)
print(sr.index)
```

 ### 1.데이터만 넣었을때  
 **2차원 배열에서 1개의 원소는 하나의 객체**

```python
df=pd.DataFrame([['1000', '선형대수',90.72],
 ['1001', 'Java',78.09],
 ['1002', '소프젝1',98.43],
 ['1003', '미적분',64.19],
 ['1004', '미디어',81.30],
 ['1005', 'Python',99.14]])
df
```

### 2.column 열 추가했을 때
**속성을 인덱스로 표시**

```python
df=pd.DataFrame([['1000', '선형대수',90.72],
 ['1001', 'Java',78.09],
 ['1002', '소프젝1',98.43],
 ['1003', '미적분',64.19],
 ['1004', '미디어',81.30],
 ['1005', 'Python',99.14]],columns=["코드","과목","점수"])
df
```

### 3.row 행 까지 완벽하게 추가하기  
**행은 객체를 열은 속성을 나타냄**

```python
df=pd.DataFrame([['1000', '선형대수',90.72],
 ['1001', 'Java',78.09],
 ['1002', '소프젝1',98.43],
 ['1003', '미적분',64.19],
 ['1004', '미디어',81.30],
 ['1005', 'Python',99.14]],columns=["코드","과목","점수"],index=[x for x in range(1,df.shape[0]+1)])
df
```

###4.딕셔너리로 생성하기
**key가 columns를 이룸, index는 따로추가**

**딕셔너리 원소 하나=하나의 열**

```python
 #딕셔너리로생성하기
data={
 '학번' :['1000', '1001', '1002', '1003', '1004', '1005'],
 '이름' :[ '선형대수', 'Java', '통계학','미적분','미디어', 'Python'],
 '점수':[90.72,78.09,98.43,64.19, 81.30,99.14]
 }
 df= pd.DataFrame(data)
 df
```

```python
data2=[["여자","남자","남자","남자","남자"],['A+','S+','B+','A','B'],[22,52,21,20,24]]
row=["조은서","임종기","배인환","황성연","배선웅"]
col=["성별","등급",'나이']
res2=pd.DataFrame(zip(*data2),index=row,columns=col)
print(res2.index.tolist())
print(res2.columns.tolist())
res2
res2.loc[["조은서","임종기"]]
```
