# Matplotlib 학습하기  
데이터를 그래프로 나타내는 패키지  
**데이터 분석 이전에 데이터이해를 위한 시각화나,데이터분석 후에 결과를 시각화
 하기 위해서 사용됩니다**  
1주차 (11월 23일)

```python
import matplotlib as mpl
import matplotlib.pyplot as plt

fig1,ax1=plt.subplots()
ax1.plot([10,20,30,40],[1,4,9,16])
fig1.suptitle("Working sheet")
ax1.set_title("Sample1")
ax1.set_xlabel("time")
ax1.set_ylabel("work")
fig1.show()
```

**실무에서는 plt.plot보다 OOP개념을 차용해서**


> **fig(도화지),ax(그래프)=plt.subplots(행,열,figsize=(도화지 크기))**  
이걸 많이 쓴다

밑에 코드는 한번에 여러 그래프를 만들고 수정하는 실습을 해보겠다.


```python
fig2,ax2=plt.subplots(2,2,figsize=(10,10))
ax2[0,0].set_title('static')
ax2[0,0].set_xlabel("day")
ax2[0,0].set_ylabel("score")
ax2[0,1].set_title("liner graph")
ax2[0,1].set_xlabel("day")
ax2[0,1].set_ylabel("score")
ax2[1,0].set_title("Java")
ax2[1,0].set_xlabel("day")
ax2[1,0].set_ylabel("score")
ax2[1,1].set_title("Python")
ax2[1,1].set_xlabel("day")
ax2[1,1].set_ylabel("score")
fig2.suptitle("Total Score")
ax2[0,0].plot([10,20,30,40],[1,4,9,16])
ax2[0,0].plot([10,20,30,40],[2,4,8,2])
ax2[0,0].legend(['score1','score2'])
ax2[0,1].plot([10,20,30,40],[2,4,14,16])
ax2[1,0].plot([10,20,30,40],[3,7,9,15])
ax2[1,1].plot([10,20,30,40],[6,16,9,3])
fig2.show()
```
